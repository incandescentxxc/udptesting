```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
log = open('udp5M1K_2.log',"r")
lines = log.readlines()
arr = []
ylabel = []
for line in lines:
    arr.append(int(line.strip()))

for i in range(len(arr)):
    ylabel.append(1)
log.close()
plt.figure(figsize=(50,30))
plt.plot(arr,ylabel,"ro")
plt.axvline(x=50000)
plt.axvline(x=100000)
plt.axvline(x=150000)
print("Total packets number " + str(len(arr)))
plt.show()
```
